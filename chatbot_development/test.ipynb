```python
import sqlite3
from langgraph.checkpoint.sqlite import SqliteSaver

# Connect to a SQLite database (it will be created if it doesn't exist)
# Use check_same_thread=False for multi-threading environments if needed
conn = sqlite3.connect("checkpoints.sqlite", check_same_thread=False)

# Initialize the SqliteSaver
memory = SqliteSaver(conn)

# You can then use 'memory' as the checkpointer when compiling your LangGraph workflow
# Example: graph = builder.compile(checkpointer=memory)
```
